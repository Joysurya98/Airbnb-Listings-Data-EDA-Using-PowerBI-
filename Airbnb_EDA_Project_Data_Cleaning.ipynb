## **Data Dupilcate and Missing value Cases**

```python
#store in temporary dataframes
c_df=chicago_df
o_df=new_orleans_df
```

### *Chicago Airbnb dataset*

```python
c_df.info()
```

```python
c_df['id'].nunique()
```

```python
c_df['name'].nunique()
```

```python
c_df[['name','host_id']].nunique()
```

```python
c_df['name'].value_counts()
```

```python
c_df.loc[c_df['name']=="Traveler's Dream - 1 bed in a shared bedroom"]
#these are duplicate values that needs to be deleted
```

```python
#delete duplicate place names
c_df=c_df.drop_duplicates(subset='name')
```

```python
c_df['host_name'].value_counts()
```

```python
c_df.loc[c_df['host_name']=="Blueground"]
#insight: we have same host but for different airbnbs in the same area
```

```python
#check for null values
c_df.isnull().mean() * 100
```

```python
#drop neighborhood group
c_df=c_df.drop(columns=['neighbourhood_group'])
```

```python
#drop license column because it has 30% null values and has no use to the problem statement
c_df=c_df.drop(columns=['license'])
```

```python
#fill null values in price column with median price calculated on the basis of same room type and neighborhood
c_df['price'] = (
    c_df.groupby(['neighbourhood', 'room_type'])['price']
      .transform(lambda x: x.fillna(x.median()))
)
```

```python
#for last_review (date column), null values suggest the airbnb is not reviewed yet
#therefore, keeping the null as it is for date analysis is the best option
```

```python
#for reviews_per_month, we can consider these nulls as 0
c_df['reviews_per_month'] = c_df['reviews_per_month'].fillna(0)
```
